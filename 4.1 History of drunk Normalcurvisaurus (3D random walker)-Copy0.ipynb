### 4. Norman class:

I would like to start with class, responsible for storing information about Norman position, making a step and calculating covered distance:

```python
N = 100     #size of space
L = 10      #numbers of Normans
dim = 3     #dimensions 3D

def PBC(xyz):
    """
    Applying boundary condition
    
    Keyword arguments:
    xyz - position vecotor of Norman
    
    Returns postition vector with PBC applied
    """
    return xyz % N

class Norman:
    """
    Class responsible for Norman behaviour, simulating one ralization of path
    
    Constructor arguments:
    xyz0 = tuple (x0, y0, z0) initial position of Norman
    dxdydz = displacement ->default (0,0,0)
    """
    def __init__(self,xyz0):
        self.r = [np.array((xyz0[0],xyz0[1],xyz0[2]))]    #list of position vectors
        self.create_steps()     #method creating possible movement possieble to create generala formula for n-dimensions
        self.dxdydz = [np.array((0,0,0))]                 #list of displecement vetors
        
    def move(self): 
        """
        Moves Norman with randomly chosen step to the next point according
        """
        step = self.pos_step[np.random.randint(6)]  #random direction
        self.r.append(step+self.r[-1])              #adding current position vector
        self.dxdydz.append(step+self.dxdydz[-1])    #adding current displacement vector
        self.r[-1] = PBC(self.r[-1])                #applying boundary conditions (defined above class)
        
    def create_steps(self):
        """
        Creates list of possible steps in case of 3d 6 possibilities
        """
        self.pos_step = np.array([[1,0,0],[0,1,0],[0,0,1],[-1,0,0],[0,-1,0],[0,0,-1]])  #list of possible moves
        
```

### 5. Main Code:

```python
def gen_pos(N):
    """
    Method generating all possible combination of indeces of 3dimensional discrete space
    Returns list of combination without repetitions (set).
    """
    return list(set(combinations(list(range(N))+list(range(N))+list(range(N)),3)))

normans = [] #initialization of list of normans
pos_positions = np.array(gen_pos(N))  #possible initial positions
indexes = np.random.choice(len(pos_positions), L, replace=False)  #picking different initial positions for each norman

#initializing loop
for pos in indexes:  
    normans.append(Norman(pos_positions[pos]))

T_end = 100 #numer of steps 
time = range(T_end) #list of steps

#update loop
for dt in time:  
    for norman in normans:
        norman.move()
```

### 6. Visualization and variance:

Since we are expecting average displacement to be equal to $<\Delta x> = 0$, so according to definition of variance: $$Var(x) = <\Delta x ^2> - <\Delta x>^2 = <\Delta x ^2> - 0 = <\Delta x ^2>$$

```python
#initializing figure
fig = plt.figure(1,facecolor='white')
ax = fig.add_subplot(111, projection='3d')
var = []

#drawing graph for every norman
for norman in normans:
    xyz = np.array(norman.dxdydz)
    ax.plot3D(xyz[:,0], xyz[:,1], xyz[:,2]) 
    var.append((np.sum(xyz**2, axis=1)))    #calculation of dx^2


plt.show()
fig = plt.figure(2,facecolor='white')
average = np.average(var,axis=0)            #average of dx^2
D = average[1:-1]/(2*dim*np.array(time)[1:])
plt.plot(time[1:],D)                        #plot of variance versus step
plt.show()
```
